# Matrix Arithmetik

Wir können mit Matrizen mathematische Operationen Element für Element durchführen, genau wie mit Vektoren. Schauen wir uns einige Beispiele an:

```python
mat <- matrix(1:50,byrow=TRUE,nrow=5)
mat
```

```python
# Multiplikation
2*mat
```

```python
# Division
1/mat
```

```python
# Division
mat/2
```

```python
# Potenz
mat ^ 2
```

## Vergleichsoperatoren mit Matrizen

Wir können des Weiteren Vergleichsoperatoren auf eine ganze Matrix anwenden, was uns eine Matrix logischer Werte als Ergebnis gibt.

```python
mat > 17
```

## Matrix Arithmetik mit mehrere Matrizen

Wir können außerdem Arithmetik auf mehrere Matrizen anwenden. Ein Beispiel:

```python
mat + mat
```

```python
mat / mat
```

```python
# Achtung, große Zahlen!
mat ^ mat
```

```python
mat*mat
```

## Matrix Multiplikation

Eine kurze Notiz zur Multiplikation von Matrizen in R. Die Multiplikation Element für Element können wir mit dem * Zeichen durchführen. Echte [Matrizenmultiplikation](https://de.wikipedia.org/wiki/Matrizenmultiplikation) ist allerdings etwas anderes. Falls ihr die Mathematik dahinter versteht und echte Matrizenmultiplikation mit R durchführen wollt, dann geht es folgendermaßen:

```python
mat2 <- matrix(1:9, nrow=3)
```

```python
mat2
```

```python
# Echte Matrizenmultiplikation
mat2 %*% mat2
```
